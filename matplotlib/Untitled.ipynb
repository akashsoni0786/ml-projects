```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(0,30,3)
```

```python
plt.plot(x,y)
```

```python
a=[9,7,3,2,5,1,8,4,2,8]
b=[1,4,2,5,6,6,7,8,9,7]
plt.plot(a,b,linestyle='dotted')
```

```python
plt.plot(a,b,linestyle='dashed')
```

```python
plt.plot(a,b,linestyle='dashdot',color='g',linewidth=10)
plt.plot(y,x,linestyle='dotted')
```

```python
font1 = {"Size":20,"family":"Serif","color":"g"}
plt.plot(a,b,linestyle='dashdot',color='g',linewidth=2,marker= 'o')
plt.plot(y,x,linestyle='dotted')
plt.xlabel("Akash",fontdict= font1)
plt.ylabel("Dev",fontdict= font1)
plt.title("Shailendra",fontdict=font1)

```

# Bar Graph

```python
x1= [1,2,3,4,5]
y1= [30,38,2,38,49]
plt.bar(x1,y1,width=.8)
```

```python
plt.barh(x1,y1)
```

```python
plt.barh(x1,y1,color='g')
```

```python
clr=['g','r','b','y','m']
plt.barh(x1,y1,color=clr)
```

```python
bills = 10
l = [10,2,20,29,12,30,4,2,1,12,3]
plt.hist(l,bins=bills)
```

```python
plt.scatter(a,b,linewidths=2,marker= "*")
plt.scatter(x1,y1)
```

```python
Lbl=["Apple","Mango","Grapes","Banana"]
ar = [20,10,30,50.654357]
plt.pie(ar,labels=Lbl,startangle= 90,explode=[0.2,0,0,0],shadow = 2,autopct='%1.6f%%')
plt.legend(title= "Fruits")
```
