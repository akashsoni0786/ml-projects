```python
df = pd.read_csv("C:\\Users\\AKASH SONI\\Desktop\\bangalore house price prediction OHE-data.csv")
df
```

```python
df.columns
```

```python
X= df.drop('price',axis=1)
y = df['price']
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state = 5)
print(X_train.shape)
print("/n 1",X_test)
print("/n 2 ",y_train)
print('/n 3',y_test)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test = sc.transform(X_test)
```

# ML Model Training

```python
from sklearn.linear_model import LinearRegression


lr= LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
print(lr.score(X_test,y_test)*100,"%")
```
