```python
gender = pd.get_dummies(data["Sex"])
gender
```

```python
gender = pd.get_dummies(data["Sex"],drop_first=True)
gender
```

```python
embark = pd.get_dummies(data["Embarked"])
embark
```

```python
pclasss = pd.get_dummies(data['Pclass'])
pclasss
```

```python
data = pd.concat([data, gender, embark, pclasss],axis=1)
data
```

```python
data.drop(["Name","Sex","Ticket","Cabin","Fare","Embarked"],axis=1,inplace=True)
```

```python
data.dropna(inplace=True)
data.head()
```

```python
x= data.drop('Survived',axis=1)
x= x.iloc[:,:]
y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y)

from sklearn.linear_model import LogisticRegression
logr =LogisticRegression()
logr.fit(x_train,y_train)
```

```python
pr=logr.predict(x_test)
pr
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test,pr)
```
