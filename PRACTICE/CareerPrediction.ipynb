```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data_train = pd.read_csv("C:\\Users\\AKASH SONI\\Downloads\\stud_training.csv")
data_train.head()
```

```python
data_train.isnull().sum()
```

```python
labelencoder= LabelEncoder()

data_train["Courses"]=labelencoder.fit_transform(data_train["Courses"])
```

```python
x_train = data_train.drop("Courses",axis=1).values
y_train = data_train["Courses"].values
```

```python
data_test = pd.read_csv("C:\\Users\\AKASH SONI\\Downloads\\stud_testing.csv")
data_test.head()
```

```python

data_test["Courses"]=labelencoder.fit_transform(data_test["Courses"])
```

```python
x_test = data_test.drop("Courses",axis=1).values
y_test = data_test["Courses"].values
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(x_train,y_train)
```

```python
pr=lin.predict(x_test)
pr
```
